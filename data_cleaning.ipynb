```python
# Reload original dataset
data = pd.read_csv("dataset.csv")

# Step 1: Encode
df = pd.get_dummies(data, columns=['type'], drop_first=True)

# Step 2: Drop string columns
df = df.drop(['nameOrig', 'nameDest'], axis=1)

# Step 3: Split
X = df.drop('isFraud', axis=1)
y = df['isFraud']

# Step 4: SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)

X_res, y_res = smote.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42
)
```

```python
X_train
```

```python
# random forest algorithm
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python
# decision tree algorithm
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

```python
# extra tree classifier
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_estimators=100, random_state=42)
et.fit(X_train, y_train)

y_pred_et = et.predict(X_test)
```

```python
# xg
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb.fit(X_train, y_train)

y_pred_xgb = xgb.predict(X_test)
```

```python
from sklearn.metrics import classification_report

models = {
    "Random Forest": y_pred_rf,
    "Decision Tree": y_pred_dt,
    "Extra Trees": y_pred_et,
    "XGBoost": y_pred_xgb
}

for name, pred in models.items():
    print(f"\n{name}")
    print(classification_report(y_test, pred))
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, average_precision_score

# 1. Prepare Data (including the error features from earlier)
X = data.drop(['isFraud', 'isFlaggedFraud', 'nameOrig', 'nameDest'], axis=1)
X = pd.get_dummies(X, columns=['type'], drop_first=True)
y = data['isFraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 2. Calculate the Weight for the Fraud Class
# This tells the model to care ~800x more about a fraud case than a normal one
ratio = float(np.sum(y == 0)) / np.sum(y == 1)

# 3. Initialize XGBoost
# Use 'scale_pos_weight' instead of SMOTE for a cleaner, faster pipeline
model = XGBClassifier(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.1,
    scale_pos_weight=ratio, # Critical for imbalanced data
    tree_method='hist',     # Faster training
    random_state=42
)

model.fit(X_train, y_train)

# 4. Evaluation
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print("XGBoost Results:")
print(classification_report(y_test, y_pred))
print(f"AUPRC Score: {average_precision_score(y_test, y_prob):.4f}")
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, average_precision_score, precision_recall_curve

# 1. LOAD & CLEAN
df = pd.read_csv("dataset.csv")

# 2. FEATURE ENGINEERING (The 'Secret Sauce')
# Fraudsters often empty accounts or ignore the limit, creating balance mismatches
df['errorBalanceOrig'] = df['newbalanceOrig'] + df['amount'] - df['oldbalanceOrg']
df['errorBalanceDest'] = df['oldbalanceDest'] + df['amount'] - df['newbalanceDest']

# 3. PREPARE FEATURES
# We keep 'type' as a category for CatBoost
features = ['type', 'amount', 'oldbalanceOrg', 'newbalanceOrig', 
            'oldbalanceDest', 'newbalanceDest', 'errorBalanceOrig', 'errorBalanceDest']
X = df[features]
y = df['isFraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 4. TRAIN CATBOOST
# auto_class_weights='Balanced' is better than SMOTE here because it 
# doesn't create "fake" data, it just makes the model pay more attention to real fraud.
model = CatBoostClassifier(
    iterations=1000,
    depth=8,
    learning_rate=0.05,
    loss_function='Logloss',
    auto_class_weights='Balanced',
    verbose=100,
    cat_features=['type'] # CatBoost handles strings automatically
)

model.fit(X_train, y_train, eval_set=(X_test, y_test), early_stopping_rounds=50)

# 5. FIND THE PERFECT THRESHOLD (Optimizing Precision)
# Instead of 0.5, we find the threshold that gives us the best F1-Score
y_probs = model.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_probs)

# Find the threshold that maximizes F1
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
best_threshold = thresholds[np.argmax(f1_scores)]

# 6. FINAL EVALUATION
y_pred_best = (y_probs >= best_threshold).astype(int)

print(f"Best Threshold Found: {best_threshold:.4f}")
print("\n--- Final 'Perfect' Model Results ---")
print(classification_report(y_test, y_pred_best))
print(f"AUPRC Score: {average_precision_score(y_test, y_probs):.4f}")
```
